```python
from astropy import constants as const
import astropy.units as u

def mass_function(period, asini):
    #return (4*np.pi**2) / (1*const.G) * ((asini)**3) / ((period).to(u.s)**2)
    si = (4*np.pi**2 * (1*const.c**3)) / (1*const.G) * 1 / (period.to(u.s)**2) * (asini**3)
    return si.to(u.M_sun)
```

```python
K, P, a1 = np.loadtxt('paper/data/pb1_binaries.txt', usecols=(0, 1, 4), skiprows=1).T
```

```python
plt.figure(figsize=mnras_size(240, square=False))

asinis = np.linspace(1,1500,1000) * u.s
periods = np.linspace(5,5000,1000) * u.d
X, Y = np.meshgrid(periods, asinis)
mfs = mass_function(X, Y)

plt.contourf(X, Y, np.log(mfs.value), levels=50, cmap='coolwarm')
cbar = plt.colorbar(label=r'$Log$ mass function ($M_\odot$)')
#plt.scatter(840, 7.1, c='green')
plt.scatter(P, a1, marker='.', facecolors='none', edgecolors='black', alpha=0.5, linewidth=0.5, rasterized=True)
#plt.plot(P, a1, marker='.', facecolors='none', edgecolors='black', alpha=1, linewidth=0.5)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Orbital period (d)')
plt.ylabel(r'a$\sin{i}$ (s)')

poss = [(a1 < 10) & (P > 100)]

cbar.ax.tick_params(labelsize=7)

plt.savefig(overleaf_path + 'known_binaries.pdf', dpi=300, bbox_inches='tight', pad_inches=0)
```
